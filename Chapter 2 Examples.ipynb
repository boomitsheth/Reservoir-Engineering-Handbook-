# Example 2.1 Calculation of Volume of Gas Using Ideal Gas Behaviour 

```python
"Question is to calculate the volume of 3pounds of N-Butane gas using ideal gas behaviour"

#known 
mass = 58.123 #lbs
temp = 120 #Fdegrees 
pressure = 60 #psia
m =3 #lbs
R = 10.73

V= (m/mass)*(R*(temp+460)/pressure) #ft3

print("Volume of Gas using Ideal Gas Behaviour is:", V,"ft3")
```

# Example 2.2 Calculation of Density of N-Butane

```python
"Question is to calculate the density of N-Butane"

den = m/V

print("Density of N-Butane is:", den,"lb/ft3")
```

# Example 2.18 Using Standing Correlation, Estimate Gas Solubility at Pb  

```python
table= pd.read_csv("F:\Tarek Ahmed Reservoir Engineering Data\Chapter 2 Reservoir Fluid Properties\example2.18.csv")
table
```

```python
#T=Reservoir Temperature
#Pb=Bubble Point Pressure
#Bo=Oil Formation Volume Factor
#Psep=Sperator Pressure
#Tsep=separator Temperature
#Co=Isothermal compressibility coeffeicient of the oil 
```

```python
table['x']= (0.0125*table['API'].astype(float)) - 0.00091*table['T'].astype(float) 
table
```

```python
table['10^x'] = pow(10,table['x'])
table
```

```python
table['Predicted Rs']= table['SG'] * pow((((table['Pb'].astype(float)/18.2)+1.4) * table['10^x'] ),1.2048) 
table 
#Standing Corelation used to calculate the predicted Solubility 
```

```python
#to calculate the absolute average error in prediction of solubility at bubble point pressure
table['%error'] = (table['Predicted Rs'] - table['Rs']) * 100 / table['Predicted Rs']
table
```
