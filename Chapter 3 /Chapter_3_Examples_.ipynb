# Example 3.2  Big Butte Crude Oil System

```python
table['Y-Function\n(B)'] = table['Y-Function\n(B)'].fillna(0)
#plot Y variable vs pressure on cartesian plot 
x = table['Pressure']
y = table['Y-Function\n(B)']
Polynomial = np.polynomial.Polynomial

cmin, cmax = min(x), max(x)
pfit, stats = Polynomial.fit(x, y, 1, full=True, window=(cmin, cmax),domain=(cmin, cmax))

pylab.plot(x, y, 'o', color='k')
pylab.plot(x, pfit(x), color='k')
pylab.xlabel('Pressure')
pylab.ylabel('Y variable')
pylab.show()
```
